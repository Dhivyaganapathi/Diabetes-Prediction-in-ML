```python
#TRAIN AND TEST
```

```python
from sklearn.model_selection import train_test_split #split the data and label
```

```python
x = df.iloc[:,df.columns!='Outcome']  #data
y = df.iloc[:,df.columns=='Outcome']  #Label / outcome
```

```python
print(x)
```

```python
print(y)
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.2)  #20% test data
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
#ALGORITHM
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
```

```python
model.fit(xtrain,ytrain.values.ravel()) #train the algorithm
RandomForestClassifier()
```

```python
predict_output = model.predict(xtest) #to test the algorithm
print(predict_output)
```

```python
accuracy = model.score(xtest,ytest)
```

```python
print("The accuracy score for RandomForest(RF) : ",accuracy)
```
